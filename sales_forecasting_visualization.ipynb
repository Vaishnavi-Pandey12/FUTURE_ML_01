# Sales Forecasting Data Visualization Report

This notebook presents a clean and structured sales visualization workflow for a forecasting project. It is designed for Binder execution using only **pandas**, **numpy**, and **matplotlib**.

The analysis covers:
- data loading and inspection,
- historical and forecast trend visualization,
- seasonality analysis,
- sales distribution,
- moving average trend,
- and a final dashboard summary.

## 1. Import Libraries

In this section, we import the required Python libraries and define default plotting settings for readability.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Global plotting style settings
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.grid'] = True
```

## 2. Load Dataset

We load `sales_forecast_dataset.csv`, convert the `Date` column to datetime format, sort by date, and display a preview.

```python
# Load dataset
df = pd.read_csv('sales_forecast_dataset.csv')

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Ensure chronological order
df = df.sort_values('Date').reset_index(drop=True)

# Preview first rows
df.head()
```

## 3. Data Inspection

We inspect structure, summary statistics, and missing values to understand data quality before visualization.

```python
print('Dataset Info:')
print('-' * 40)
print(df.info())

print('\nSummary Statistics:')
print('-' * 40)
print(df.describe(include='all', datetime_is_numeric=True))

print('\nMissing Values:')
print('-' * 40)
print(df.isnull().sum())
```

## 4. Monthly Sales Trend Visualization

This line chart shows sales over time with separate lines for `Actual` and `Forecast` records.

```python
actual = df[df['Type'] == 'Actual']
forecast = df[df['Type'] == 'Forecast']

plt.figure(figsize=(10,5))
plt.plot(actual['Date'], actual['Sales'], label='Actual Sales', color='tab:blue', linewidth=2)
plt.plot(forecast['Date'], forecast['Sales'], label='Forecast Sales', color='tab:orange', linewidth=2)

plt.title('Monthly Sales Trend: Actual vs Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 5. Actual vs Forecast Comparison

This focused comparison highlights the transition from historical data to forecasted values.

```python
plt.figure(figsize=(10,5))
plt.plot(actual['Date'], actual['Sales'], marker='o', label='Actual Sales', color='tab:green')
plt.plot(forecast['Date'], forecast['Sales'], marker='o', linestyle='--', label='Forecast Sales', color='tab:red')

plt.title('Actual vs Forecast Sales Comparison')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 6. Monthly Seasonality Analysis

We extract month values from dates and compute average sales per month to reveal seasonal patterns.

```python
df['Month'] = df['Date'].dt.month
monthly_avg = df.groupby('Month', as_index=False)['Sales'].mean()

plt.figure(figsize=(10,5))
plt.bar(monthly_avg['Month'], monthly_avg['Sales'], color='mediumpurple')

plt.title('Average Sales by Month (Seasonality)')
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.xticks(np.arange(1, 13, 1))
plt.tight_layout()
plt.show()
```

## 7. Sales Distribution

A histogram helps visualize how sales values are distributed across the dataset.

```python
plt.figure(figsize=(10,5))
plt.hist(df['Sales'], bins=15, color='steelblue', edgecolor='black', alpha=0.8)

plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## 8. Rolling Average Trend

A 3-month moving average smooths short-term fluctuations and highlights broader trends.

```python
actual_ma = actual.copy()
actual_ma['Sales_3M_MA'] = actual_ma['Sales'].rolling(window=3).mean()

plt.figure(figsize=(10,5))
plt.plot(actual_ma['Date'], actual_ma['Sales'], label='Actual Sales', color='tab:blue', alpha=0.7)
plt.plot(actual_ma['Date'], actual_ma['Sales_3M_MA'], label='3-Month Moving Average', color='tab:orange', linewidth=2)

plt.title('Actual Sales with 3-Month Moving Average')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 9. Forecast Visualization

This chart emphasizes the forecast period by shading the region where predictions begin.

```python
plt.figure(figsize=(10,5))
plt.plot(actual['Date'], actual['Sales'], label='Historical Sales', color='tab:blue', linewidth=2)
plt.plot(forecast['Date'], forecast['Sales'], label='Predicted Sales', color='tab:orange', linewidth=2)

if not forecast.empty:
    forecast_start = forecast['Date'].min()
    plt.axvspan(forecast_start, df['Date'].max(), color='orange', alpha=0.15, label='Forecast Region')

plt.title('Historical vs Forecast Sales with Highlighted Forecast Region')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 10. Final Visualization Dashboard Section

We combine key views into a compact dashboard for reporting and presentation.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Panel 1: Actual vs Forecast Trend
axes[0, 0].plot(actual['Date'], actual['Sales'], label='Actual', color='tab:blue')
axes[0, 0].plot(forecast['Date'], forecast['Sales'], label='Forecast', color='tab:orange')
axes[0, 0].set_title('Trend: Actual vs Forecast')
axes[0, 0].set_xlabel('Date')
axes[0, 0].set_ylabel('Sales')
axes[0, 0].legend()
axes[0, 0].tick_params(axis='x', rotation=45)

# Panel 2: Seasonality
axes[0, 1].bar(monthly_avg['Month'], monthly_avg['Sales'], color='mediumpurple')
axes[0, 1].set_title('Monthly Seasonality')
axes[0, 1].set_xlabel('Month')
axes[0, 1].set_ylabel('Average Sales')
axes[0, 1].set_xticks(np.arange(1, 13, 1))

# Panel 3: Distribution
axes[1, 0].hist(df['Sales'], bins=15, color='steelblue', edgecolor='black', alpha=0.8)
axes[1, 0].set_title('Sales Distribution')
axes[1, 0].set_xlabel('Sales')
axes[1, 0].set_ylabel('Frequency')

# Panel 4: Moving Average
axes[1, 1].plot(actual_ma['Date'], actual_ma['Sales'], label='Actual', color='tab:blue', alpha=0.7)
axes[1, 1].plot(actual_ma['Date'], actual_ma['Sales_3M_MA'], label='3M MA', color='tab:orange', linewidth=2)
axes[1, 1].set_title('3-Month Moving Average')
axes[1, 1].set_xlabel('Date')
axes[1, 1].set_ylabel('Sales')
axes[1, 1].legend()
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
